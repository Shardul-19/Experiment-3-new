```python
marks=(78,85,90,66,88)
maximum_marks=max(marks)
print(maximum_marks)
minimum_marks=min(marks)
print(minimum_marks)
total_marks=sum(marks)
print(total_marks)
print(len(marks))
print(sum(marks)/len(marks))
# Tuples do not have a .sorted() or .sort() method as they are immutable.
# Use the built-in sorted() function, which returns a new sorted list.
print("Sorted marks (ascending):", sorted(marks))
print("Marks (original tuple, unchanged):", marks)
print("Sorted marks (descending):", sorted(marks, reverse=True))
```
